```python

class VideoNode:
    def __init__(self, title):
        self.title = title
        self.next = None
        self.prev = None


class VideoPlaylist:
    def __init__(self):
        self.head = None
        self.current = None


    def add_video(self, title):
        new_video = VideoNode(title)

        if self.head is None:
            self.head = new_video
            self.current = new_video
        else:
            temp = self.head
            while temp.next:
                temp = temp.next
            temp.next = new_video
            new_video.prev = temp


    def show_playlist(self):
        temp = self.head
        print("\nPlaylist Video:")
        while temp:
            print("-", temp.title)
            temp = temp.next


    def next_video(self):
        if self.current and self.current.next:
            self.current = self.current.next
            print("Sekarang memutar:", self.current.title)
        else:
            print("Sudah di video terakhir!")


    def prev_video(self):
        if self.current and self.current.prev:
            self.current = self.current.prev
            print("Sekarang memutar:", self.current.title)
        else:
            print("Sudah di video pertama!")


    def remove_video(self, title):
        temp = self.head

        while temp:
            if temp.title == title:
                if temp.prev is None:
                    self.head = temp.next
                    if self.head:
                        self.head.prev = None
                else:
                    temp.prev.next = temp.next
                    if temp.next:
                        temp.next.prev = temp.prev

                print(f"Video '{title}' berhasil dihapus")
                return
            temp = temp.next




if __name__ == "__main__":
    playlist = VideoPlaylist()


    playlist.add_video("Video A")
    playlist.add_video("Video B")
    playlist.add_video("Video C")


    playlist.show_playlist()


    print("\nMulai dari:", playlist.current.title)
    playlist.next_video()
    playlist.next_video()
    playlist.prev_video()


    playlist.remove_video("Video B")


    playlist.show_playlist()
```
